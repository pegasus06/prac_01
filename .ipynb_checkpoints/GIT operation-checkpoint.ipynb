## preface

**feature:**Free, open source distributed version control system that can quickly handle projects from small to large and manage codes.<br/>**advantage**:It can assign different permissions.<br/>**disadvantages**:If the central server goes down,no one can submit updates.<br/>

**common commands**<br/>
git config --global user.name 用户名 ----设置用户签名（区分不同操作者的身份）<br/>git config --global user.email 邮箱号 -----设置用户签名<br/>git init -----初始化本地库<br/>git status -----查看本地库状态<br/>git add 文件名 -----添加到暂存区<br/>git rm --cached 文件名  -----删除暂存区的文件<br/>git commit -m "日志信息"  -----提交到本地库<br/>git reflog -----查看日志信息<br/>git reset --hard 版本号 -----版本穿梭<br/>git flog -----查看详细日志信息<br/><br/>git branch 分支名 -----创建分支<br/>git branch -v -----查看分支<br/>git branch -----查看现在所在的分支<br/>git checkout 分支名 -----切换分支<br/>git merge 分支名 -----把指定的分支合并到当前分支上<br/>git rebase master --从master拉取最新的代码<br/><br/>git remote add origin 仓库地址 -----连接远程仓库<br/>git remote rm origin -----去除连接的远程库<br/>git pull --rebase origin master -----同步仓库内容到master分支<br/>git push -u origin 分支名 -----上传到远程仓库<br/>git pull origin 分支名 -----从远程仓库拉取文件<br/>git clone 仓库地址 -----克隆远程仓库<br/>


**Ps:When you finish processing conflicting files, you cannot add a file name to the submitted file**<br/>**When you first use it, you need to clone the remote library and synchronize the content**
