## 变量及其赋值

### 不可变数据类型
a = 1<br/>
b = a<br/>
a = a + 1<br/>
将1赋值给a，即a指向1这个对象，b=a表示b也指向1这个对象（python的对象可被多个变量引用）,a=a+1,是表示重新创建一个值为2的对象，并让a指向它，b仍然指向1这个对象。<br/>对于不可变对象（字符串，整型，元祖等等），所有指向该对象的变量的值总是一样的，也不会改变。但是通过某些操作（+= 等等）更新不可变对象的值时，会返回一个新的对象。


### 可变数据类型

```python
list1=[1,2,3,4]
list2=list1
list1.append(5)
print(list1)
print(list2)
```

list1和list2同时指向[1,2,3,4]这个对象,由于列表可变，append不会创建新的列表，因为同时指向，所以会同时发生变化。

### 变量的删除
python中变量可以删除，但是对象不可以删除，比如：<br/>l=[1,2,3]<br/>del l<br/>del删除l这个变量，但是对象[1,2,3]仍然存在<br/>python自带的垃圾回收会跟踪对象的引用，若对象没有变量引用，内存便会回收。

## python函数的参数传递
参数传递是引用传递（pass by object reference)，Python 里所有的数据类型都是对象，所以参数传递时，只是让新变量与原变量指向相同的对象

```python
def my_func1(b):
	b = 2
 
a = 1
my_func1(a)
print(my_func1(a))
```

```python
def my_func2(b):
	b = 2
	return b #有返回值
 
a = 1
a = my_func2(a)
print(a)
```

### *可变对象的参数传递*

```python
def my_func(l2):   #影响到所有指向它的变量
    l2.append(5)  #l1和l2先都指向[1,2,3,4],由于列表可变，添加新元素，值都会变
l1=[1,2,3,4]
my_func(l1)
print(l1)
```

**<mark>！！！只要传递值中出现了= 赋值号，就是改变了传递的指针引用</mark>**

```python
#case1
def myfunc4(l2):
    l2=l2+[4]   #表示创建了一个新列表
l1=[1,2,3]
myfunc4(l1)
print(l1)
```

```python
#case2
def myfunc(l2):
    l2=l2+[4]   #表示创建了一个新列表
    return l2
l1=[1,2,3]
l1=myfunc4(l1)
print(l1)
```

Q1:下面的代码中, l1、l2 和 l3 都指向同一个对象吗？
l1 = [1, 2, 3]
l2 = [1, 2, 3]
l3 = l2
A:列表是可变对象，每创建一个列表，都会重新分配内存，因此，L1和L2不是指向同一个对象，而l3=l2，让l3指向l2的对象。
